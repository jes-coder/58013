```python
#VECTOR OPERATIONS
#(Supplementary Activity)

import numpy as np

# Define the vectors
A = np.array([-0.4, 4.3, -0.6])
B = np.array([-0.2, 0.21, 0])
C = np.array([-0.2, 2.1, -1.5])

# Vector addition
addition_result = A + B

# Vector subtraction
subtraction_result = A - C

# Dot product
dot_product_result = np.dot(A, B)

# Cross product
cross_product_result = np.cross(A, C)

# Print results
print("Vector Addition (A + B):", addition_result)
print("Vector Subtraction (A - C):", subtraction_result)
print("Dot Product (A . B):", dot_product_result)
print("Cross Product (A x C):", cross_product_result)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the cross product result
cross_product_result = np.array([-5.19, -0.48, 0.02])

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Add a vector from the origin to the cross product result
ax.quiver(0, 0, 0, cross_product_result[0], cross_product_result[1], cross_product_result[2], color='r')

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set limits
ax.set_xlim([-6, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([0, 1])

# Show plot
plt.show()
```
